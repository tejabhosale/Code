```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
#Create Datapoints
X = np.array([
    [1, 2],
    [2, 2],
    [2, 3],
    [8, 7],
    [8, 8],
    [25, 80],
    [30, 75]
])
#Standardize the data (Optional but recommended)
X = StandardScaler().fit_transform(X)

#Apply DBSCAN
dbscan = DBSCAN(eps = 2, min_samples = 3)  #eps = maximum distance between 2 samples
labels = dbscan.fit_predict(X)

#Print the labels assigned by DBSCAN
print("DBSCAN Lables : ", labels)

#Vizualize the clustering
colors = ['purple', 'purple', 'red', 'green', 'orange', 'magenta']

for i in range(len(labels)) :
  plt.scatter(X[i , 0], X[i, 1], color = colors[labels[i]], s = 50)
plt.title('DBSCAN Clustering')
plt.xlabel("Feature 1")
plt.ylabel('Feature 2')
plt.show()
```
